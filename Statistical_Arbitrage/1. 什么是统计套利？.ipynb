## 什么是统计套利？
统计套利是根据价差同时买入与卖出一样或极为相似的品种。

## 统计套利有哪些策略？
- 跨市场套利： 不同市场的相同或相似资产
- 期现套利： 期货与现货市场
- 跨期套利： 同一品种不同月份期限的合约
- 跨品种套利： 品种不同但高度相关
- 期权套利： 蝶式鹰式

## 什么是价差与正反向套利？
统计套利中的价差指的是两相似品种的价格差。

正向套利： 买入近期，卖出远期； 买大豆，卖豆油。

反向套利： 卖出近期，买入远期； 卖大豆，买豆油。

```python
#寻找银行类股票的套利机会
import pandas as pd
import tushare as ts

codes = ['000001', '002142', '002807', '600000', '600015', 
         '600016', '600036', '600908', '600919', '600926',
         '601009', '601128', '601166', '601169', '601229',
        '601288', '601328', '601398', '601818', '601939',
        '601988', '601997', '601998', '603323']

stocks_dict = {}
for c in codes:
    stock = ts.get_k_data(c, start='2010-01-01', ktype='D', autype='qfq')
    stock.index = pd.to_datetime(stock['date'], format='%Y-%m-%d')
    stock.pop('date')
    stocks_dict[c] = stock

pn = pd.Panel(stocks_dict)

returns = pn.minor_xs('close').pct_change()[1:]
corr_r = returns.corr()
print corr_r.nlargest(5, columns='601398')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(corr_r, square=True)
plt.show()
```

```python
#选择601398，601288
import numpy as np

close = pn.minor_xs('close')
data = pd.concat([close['601398'], close['601288']], axis=1)
data['P_diff'] = data.iloc[:,0]-data.iloc[:,1]
# print data.P_diff
diff_mean = data.P_diff[:-400].mean()
print diff_mean
diff_std = data.P_diff[:-400].std()

plt.plot(data.P_diff)
plt.hlines(diff_mean, data.index[0], data.index[-400])
plt.hlines(diff_mean+2*diff_std, data.index[0], data.index[-400], colors='r')
plt.hlines(diff_mean-2*diff_std, data.index[0], data.index[-400], colors='r')
plt.show()
```

## 如何执行价差套利策略？
1. 价差区间超买超卖交易（买弱卖强）
    - 在上安全边际买入弱的卖出强的
2. 反向交易（买强卖弱）
    - 在上穿均值后买入强的卖出弱的
3. 在更强的位置重复加仓
    - 在穿越安全边际后继续做空价差
